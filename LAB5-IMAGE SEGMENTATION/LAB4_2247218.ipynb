```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def resize_nearest_neighbor(image, scale):
    height, width = image.shape[:2]
    new_height = int(height * scale)
    new_width = int(width * scale)
    resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)
    return resized_image

def resize_bilinear(image, scale):
    height, width = image.shape[:2]
    new_height = int(height * scale)
    new_width = int(width * scale)
    resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
    return resized_image

# Load two images (replace with your image filenames)
image1 = cv2.imread('pred1.jpg')
image2 = cv2.imread('pred2.jpg')

# Check if images are loaded successfully
if image1 is None or image2 is None:
    print("Error: Unable to load images.")
    exit()

# Specify the scaling factor for image sampling
scale_factor = 0.2  # Change this value as needed

# Perform image sampling using nearest neighbor interpolation
resized_image1_nearest = resize_nearest_neighbor(image1, scale_factor)
resized_image2_nearest = resize_nearest_neighbor(image2, scale_factor)

# Perform image sampling using bilinear interpolation
resized_image1_bilinear = resize_bilinear(image1, scale_factor)
resized_image2_bilinear = resize_bilinear(image2, scale_factor)




```

```python
# plt.figure(figsize=(12, 6))
# plt.subplot(2, 2, 1)
# cv2_imshow(image1)
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(image1, cmap='gray')
plt.title('PET Image')

plt.subplot(2, 2, 2)
plt.imshow(resized_image1_nearest, cmap='gray')
plt.title('Resized Image')

plt.subplot(2, 2, 3)
plt.imshow(resized_image1_bilinear, cmap='gray')
plt.title('Resized Image1 Bilinear')

plt.subplot(2, 2, 4)
plt.imshow(resized_image2_bilinear, cmap='gray')
plt.title('Resized Image2 Bilinear')

plt.tight_layout()
plt.show()
```

## Conclusion

In this code example, we performed image resizing on two input images using two different interpolation methods: nearest neighbor and bilinear interpolation. Here are the key observations:

1. **Image Loading:** We successfully loaded two images, 'pred1.jpg' and 'pred2.jpg,' using OpenCV. It's important to ensure that images are loaded correctly before further processing.

2. **Scaling Factor:** We defined a `scale_factor` variable, which determined the degree of resizing applied to the images. This factor can be adjusted to control the amount of scaling.

3. **Resizing Techniques:**
   - **Nearest Neighbor Interpolation:** This method resizes images using the pixel value of the nearest neighbor. It can result in blocky or pixelated artifacts when upscaling.
   - **Bilinear Interpolation:** Bilinear interpolation takes into account the values of neighboring pixels to calculate the interpolated pixel value. It generally produces smoother results compared to nearest neighbor interpolation.

4. **Visualization:** We used the `matplotlib.pyplot` library to create a 2x2 grid of subplots for visual comparison:
   - The first subplot displayed the original PET image.
   - The second subplot displayed the image resized with nearest neighbor interpolation.
   - The third subplot displayed the image resized with bilinear interpolation.
   - The fourth subplot displayed another image resized with bilinear interpolation for comparison.

5. **Conclusion:** This code serves as a practical demonstration of image resizing and the visual differences between nearest neighbor and bilinear interpolation methods. Depending on the specific use case and image content, one interpolation method may be more suitable than the other.
